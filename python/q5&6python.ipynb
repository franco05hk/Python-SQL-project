```python
df = pd.DataFrame(data)
df.drop(columns=['category'], inplace=True)
```

```python
df
```

```python
def categorize_country(country):
    if country in ['United States', 'Canada', 'Mexico']:
        return 'Countries_of_freedom'
    elif country in ['China', 'Russia', 'North Korea']:
        return 'Countries_of_autocracy'
    elif country in ['Brazil', 'Kazakhstan', 'Peru']:
        return 'Developing_countries'
    elif country in ['Australia', 'France', 'United Kingdom']:
        return 'Commonwealth_selection'
    elif country in ['Egypt', 'Iraq', 'United Arab Emirates']:
        return 'Arab_league'
    elif country in ['Afghanistan', 'Ukraine']:
        return 'War_zone'
    else:
        return 'Unclassified'
pd.set_option('display.max_rows', None)
# Apply the categorize_country function to create the 'Category' column
df['Category'] = df['country'].apply(categorize_country)

print(df)
```

```python
category_counts = df['Category'].value_counts()

print(category_counts)
```
