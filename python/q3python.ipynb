```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('q3.csv')
```

```python
data.head
```

```python
data.describe
```

```python
df = pd.DataFrame(data)

# Sort the DataFrame by 'sum' in descending order for a better visualization
df = df.sort_values(by='sum', ascending=False)

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(df['country'], df['sum'])
plt.xlabel('Country')
plt.ylabel('Sum')
plt.title('Sum of Values by Country')
plt.xticks(rotation=45)  # Rotate country labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```

```python
top_20_df = df.sort_values(by='sum', ascending=False).head(20).copy()

# Create a bar plot for the top 20 countries
plt.figure(figsize=(10, 6))
plt.bar(top_20_df['country'], top_20_df['sum'])
plt.xlabel('Country')
plt.ylabel('Sum')
plt.title('Top 20 Countries by Sum of Values')
plt.xticks(rotation=45)  # Rotate country labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```
